```python
import math
time_step = 0
valid_obstacle_ids = list()
config.vehicle.ego_id = 5
config.print_configuration_summary()
evaluator = ET(config)
obstacles = evaluator.sce.dynamic_obstacles
for i in obstacles:
    print(i.obstacle_id)
    other_veh_id = i.obstacle_id
    psd = evaluator.compute(other_veh_id, time_step)
    if psd != 0 and psd != math.inf:
        valid_obstacle_ids.append(i)
```

```python
for i in valid_obstacle_ids:
    evaluator.compute(i.obstacle_id, 0)
    config.debug.save_plots = False
    evaluator.visualize()
```
